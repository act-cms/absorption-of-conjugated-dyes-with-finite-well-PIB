```python
from IPython.display import display, HTML

def collapsible_quiz_display(encoded_json_path='quiz_encoded.json'):
    # Inject collapsible markdown toggles
    display(HTML('''
    <details open>
    <summary style="cursor:pointer; font-weight:bold; font-size:16px;">Show/Hide Code</summary>
    '''))

    # Your code block — decode and display quiz
    import json, base64
    from jupyterquiz import display_quiz

    with open(encoded_json_path, 'r') as f:
        data = json.load(f)

    quiz_b64 = data["data"]
    quiz_json = base64.b64decode(quiz_b64).decode("utf-8")
    quiz_data = json.loads(quiz_json)

    display_quiz(quiz_data)

    # Close the collapsible block
    display(HTML('''
    </details>
    '''))
```
