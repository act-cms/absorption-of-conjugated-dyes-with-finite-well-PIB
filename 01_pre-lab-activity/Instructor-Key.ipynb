# Part 2. `If`/`Else` Statements

In programming, sometimes we want our code to **make decisions** based on certain conditions. That’s exactly what `if`/`else` statements do: they allow the program to **execute different blocks of code depending on whether a condition is true or false**.

---

### Basic Structure
```python
if condition:
    # code runs if condition is True
else:
    # code runs if condition is False
```

condition: A logical expression that evaluates to True or False.

Indentation: Python uses indentation to group code blocks. All code inside the if or else must be indented.

### Example 1: Checking a Number
```python
x = 10

if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or smaller")
```

**Explanation:** The program checks whether x > 5. Since 10 > 5 is True, it prints: "x is greater than 5" If x were 3, the condition would be False, and the else block would execute.

### Example 2: Multiple Conditions with `elif`

Sometimes we have more than two possibilities. We can use elif (short for "else if"):
```python
temperature = 25

if temperature > 30:
    print("It's really hot!")
elif temperature > 20:
    print("It's warm.")
else:
    print("It's cool.")
```

**Explanation:**
The program checks conditions in order. Only the first True condition will execute. If none are True, the else block runs.

# Part 3. For loops

In Python, a **`for` loop** is used to iterate over a sequence, such as a list, tuple, string, or range, allowing you to execute a block of code repeatedly for each item in the sequence.


## 1. Basic Syntax
```python
for variable in sequence:
    # code block
```

*   variable: A placeholder for each element in the sequence.
*   sequence: The collection (e.g., list, tuple, string) you want to iterate over.

Example:
```python
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)
```

## 2. Range
The `range()` function generates a sequence of numbers, which is often used in loops.
```python
for i in range(5):  # Generates numbers from 0 to 4
    print(i)
```

> `range(start, stop, step)` allows customization of the range.


You can get more detailed with the specification, like looping through different data types (`item in [1, 2, 3]` will give 1, 2, 3 for example).

Multidimensional lists can take advantage of loops within loops, also called **nested loops**
```python
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
```

### The += operator

See how the += operator works in a for loop and what the operator does mathematically.

````python
# s: store sum of all numbers
s = 0
n = int(input("Enter number "))

# run loop n times
# stop: n+1 (because range never include stop number in result)
for i in range(1, n + 1, 1):
    # add current number to sum variable
    s += i
    print(i,s)
print("Sum is: ", s)
````

````python
def gaussian_peak(x, peak_location, amplitude, broadening):
    """
    Generates a Gaussian peak.

    Args:
        x (numpy.ndarray): The x-axis values.
        peak_location (float): The center of the Gaussian peak.
        broadening (float): The standard deviation of the Gaussian peak (controls width).

    Returns:
        numpy.ndarray: The y-values of the Gaussian peak.
    """

    return amplitude * np.exp(-((x - peak_location)**2) / (2 * broadening**2))
````

````python
# Define the linear space between 1 and 10 with 10000 points
X=np.linspace(1,10,10000)

# Define peak data
peaks=[1,2.1,5,6.7,8,9]
amps=[3,1,2,5,0.3,2]

# Start of with a- list of zeros.
# You will use the += operation to add the contributions of each computed peak
comp_spec1=np.zeros(len(X))
comp_spec2=np.zeros(len(X))

# Define the for loop
for i in range(len(peaks)):
  comp_spec1+=gaussian_peak(X,peaks[i],amps[i],0.25)
  comp_spec2+=gaussian_peak(X,peaks[i],amps[i],0.50)
````

````python
# Plot the computed spectra with the vertical lines representing your computed spectra
plt.plot(X,comp_spec1)
plt.plot(X,comp_spec2)

plt.xlabel('x')
plt.ylabel('y')
plt.show()
````
